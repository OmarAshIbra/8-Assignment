###  Function To Generate Recommended Episodes| معادلة لتوليد حلقات مقترحة   

- Recommendation Criteria :
* Criterion 1 : Recommend the most listened-to episode in categories the user has already interacted with.
* Criterion 2 : Recommend the most listened-to episode among users of the same gender.
* Criterion 3 : Recommend the most listened-to episode in the user's country.
- Constraints :
* The recommended episodes must be unique.
* The recommended episodes must not have been listened to by the user before.

```python
### Function to Recommend Episodes
def recommend_episodes(user_id: int, listens_df: pd.DataFrame, episodes_df: pd.DataFrame, users_df: pd.DataFrame) -> list:
    # Step 1: Validate user exists in users_df
    user_row = users_df[users_df['user_id'] == user_id]
    if user_row.empty:
        raise ValueError(f"User with ID {user_id} not found in users_df")
    
    # Extract user  information
    user_gender = user_row.iloc[0]['gender']
    user_country = user_row.iloc[0]['country']
    # print(f"User {user_id} is from {user_country} and is {user_gender}.")    -- test

    # Step 2: Get episodes already listened by the user
    user_listened_episodes = listens_df[listens_df['user_id'] == user_id]['episode_id'].unique()
    # print(f"User {user_id} has listened to {len(user_listened_episodes)} episodes.")    -- test

    # Step 3: Merge listens_df with episodes_df to get episode details
    merged_df = pd.merge(listens_df, episodes_df, on='episode_id', how='inner')
    
    # display(merged_df.head())        -- test
    # print(merged_df.shape)           -- test

    # Step 4: Compute total listen time per episode
    episode_popularity = merged_df.groupby('episode_id')['duration_seconds'].sum().reset_index()
    episode_popularity.rename(columns={'duration_seconds': 'total_listens'}, inplace=True)

    # Join with episodes_df to get full episode details
    episode_details = pd.merge(episode_popularity, episodes_df, on='episode_id', how='inner')
    # display(episode_details.head())   -- test
    # print(episode_details.shape)      -- test
    

    # Exclude episodes the user has already listened to
    eligible_episodes = episode_details[~episode_details['episode_id'].isin(user_listened_episodes)]
    # display(eligible_episodes.head())     -- test
    # print(eligible_episodes.shape)        -- test

    # Step 5: Recommendation Criteria

    # 1. Most listened-to episode in categories the user has interacted with
    # Get the categories of episodes the user has listened to
    user_listen_episodes = listens_df[listens_df['user_id'] == user_id]['episode_id']
    user_listen_with_category = episodes_df[episodes_df['episode_id'].isin(user_listen_episodes)]
    user_categories = user_listen_with_category['category'].unique()
    # display(user_categories)

    category_mask = eligible_episodes['category'].isin(user_categories)
    category_candidates = eligible_episodes[category_mask]
    recommendation_1 = None
    if not category_candidates.empty:
        recommendation_1 = category_candidates.sort_values(by='total_listens', ascending=False).iloc[0]['episode_id']
        # print(f"Recommendation 1 (most listened in categories): Episode-{recommendation_1}")      -- test

    # 2. Most listened-to episode among users of the same gender
    # Get all episodes listened by users of the same gender
    same_gender_users = users_df[users_df['gender'] == user_gender]['user_id']
    gender_listen_data = listens_df[listens_df['user_id'].isin(same_gender_users)]
    gender_merged = pd.merge(gender_listen_data, episodes_df, on='episode_id')
    gender_episode_popularity = gender_merged.groupby('episode_id')['duration_seconds'].sum().reset_index()
    gender_episode_popularity.rename(columns={'duration_seconds': 'total_listens'}, inplace=True)
    gender_candidates = gender_episode_popularity[~gender_episode_popularity['episode_id'].isin(user_listened_episodes)]
    recommendation_2 = None
    if not gender_candidates.empty:
        recommendation_2 = gender_candidates.sort_values(by='total_listens', ascending=False).iloc[0]['episode_id']
        # print(f"Recommendation 2 (most listened by his Gender): Episode-{recommendation_2}")      -- test


    # 3. Most listened-to episode in the user's country
    eligible_episodes= pd.merge(eligible_episodes, listens_df[['episode_id', 'user_id']], on='episode_id', how='inner')
    eligible_episodes= pd.merge(eligible_episodes, users_df[['user_id', 'country']], on='user_id', how='inner')
    
    country_candidates = eligible_episodes[eligible_episodes['country'] == user_country]
    recommendation_3 = None
    if not country_candidates.empty:
        recommendation_3 = country_candidates.sort_values(by='total_listens', ascending=False).iloc[0]['episode_id']

    # Step 6: Ensure recommendations are unique
    recommendations = [recommendation_1, recommendation_2, recommendation_3]
    unique_recs = []
    for rec in recommendations:
        if rec is not None and rec not in unique_recs:
            unique_recs.append(rec)

    # Step 7: Fill in missing or duplicate slots using globally most listened episodes
    if len(unique_recs) < 3:
        global_top = eligible_episodes.sort_values(by='total_listens', ascending=False)['episode_id'].unique()
        for ep_id in global_top:
            if ep_id not in unique_recs:
                unique_recs.append(ep_id)
                if len(unique_recs) == 3:
                    break

    return unique_recs[:3]
```

```python
### Test the recommendation function
user_id = 7  # Replace with the desired user_id 
recommended_episodes = recommend_episodes(user_id, listens_df, episodes_df, users_df)
print(f"\nRecommended episodes for user {user_id}:")
for episode_id in recommended_episodes:
    episode_title = episodes_df[episodes_df['episode_id'] == episode_id]['title'].values[0]
    print(f"Episode ID: {episode_id}, Title: {episode_title}")
```
